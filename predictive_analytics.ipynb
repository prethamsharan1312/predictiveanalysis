# Predictive Analytics Using Historical Data
This notebook demonstrates:

- Synthetic historical dataset generation (>= 1000 rows)
- Data cleaning (missing values + outliers)
- Feature engineering (calendar features, lags, rolling stats)
- Model training & evaluation
  - Linear Regression baseline
  - Statsmodels SARIMAX time-series forecasting
- Metrics: MAE, MSE, RMSE, R²
- Actual vs Predicted comparisons and forecast visualization


```python
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from predictive_models import DatasetConfig, generate_synthetic_historical_data, handle_missing_and_outliers, feature_engineering, train_evaluate_linear_regression, train_evaluate_sarimax
from data_pipeline import make_dataset_and_features

sns.set_theme(style='whitegrid')

# Paths
BASE_DIR = os.path.abspath('..')
ARTIFACTS_DIR = os.path.join(BASE_DIR, 'artifacts')
os.makedirs(ARTIFACTS_DIR, exist_ok=True)
```

## 1) Load / Generate Dataset
Since no external dataset is provided, we generate a synthetic time-series with:

- daily timestamps
- numeric drivers
- a categorical driver
- trend, seasonality, noise
- injected missing values and outliers


```python
cfg = DatasetConfig(n_records=1500, freq='D', start_date='2018-01-01', seed=42)
raw_df = generate_synthetic_historical_data(cfg)
raw_df.head()
```

```python
raw_df.info()
raw_df.describe(include='all')
```

### Missing Values Check

```python
raw_df.isna().mean().sort_values(ascending=False)
```

## 2) Data Cleaning & Outlier Handling
We:
- fill numeric missing values with median
- fill categorical missing values with mode
- clip outliers using robust quantile bounds (1% to 99%)


```python
clean_df = handle_missing_and_outliers(raw_df)
clean_df.isna().mean().sort_values(ascending=False)
clean_df[['numeric_feature_2']].describe()
```

## 3) Feature Engineering
We create supervised learning features:
- calendar features: day_of_week, month
- lag features: target_lag_1..target_lag_14
- rolling features: rolling_mean_7, rolling_mean_14, rolling_std_7
- a lightweight target-mean encoding for categorical_feature


```python
max_lag = 14
df_feat = feature_engineering(clean_df, max_lag=max_lag)
df_feat.head()
df_feat.shape
```

## 4) Visualization: Historical Trends


```python
plt.figure(figsize=(12,4))
plt.plot(clean_df['timestamp'], clean_df['target'], linewidth=1)
plt.title('Historical Target Trend')
plt.xlabel('Date')
plt.ylabel('Target')
plt.tight_layout()
plt.show()

# Seasonal-ish view by day of week
tmp = clean_df.copy()
tmp['day_of_week'] = tmp['timestamp'].dt.dayofweek
plt.figure(figsize=(10,4))
sns.boxplot(data=tmp, x='day_of_week', y='target')
plt.title('Target Distribution by Day of Week')
plt.tight_layout()
plt.show()
```

## 5) Split into Training & Testing Sets (time ordered)
We use an 80/20 split based on time order to avoid leakage.

```python
# Training/testing happens inside the training functions
# so we only show the engineered dataframe size here.
len(df_feat), df_feat['timestamp'].min(), df_feat['timestamp'].max()
```

## 6) Model 1: Linear Regression Baseline
We predict y = target(t + horizon) using engineered supervised features at time t.

```python
horizon = 7
lin_results = train_evaluate_linear_regression(df_feat, horizon=horizon)
lin_results.keys()
print({k: lin_results[k] for k in ['MAE','MSE','RMSE','R2']})
```

```python
lin_pred_df = lin_results['predictions']
lin_pred_df.head()
```

### Actual vs Predicted (Linear Regression)

```python
plt.figure(figsize=(12,4))
plt.plot(lin_pred_df['timestamp'], lin_pred_df['actual'], label='Actual', linewidth=1)
plt.plot(lin_pred_df['timestamp'], lin_pred_df['predicted'], label='Predicted', linewidth=1)
plt.title('Actual vs Predicted (Linear Regression)')
plt.xlabel('Date')
plt.ylabel('Target')
plt.legend()
plt.tight_layout()
plt.show()
```

## 7) Model 2: Time Series Forecasting with Statsmodels SARIMAX
We forecast the univariate target series for the test window length.

```python
order = (1,1,1)
sarimax_results = train_evaluate_sarimax(df_feat, order=order)
print({k: sarimax_results[k] for k in ['MAE','MSE','RMSE','R2']})
```

```python
sarimax_pred_df = sarimax_results['predictions']
sarimax_pred_df.head()
```

### Actual vs Predicted (SARIMAX)

```python
plt.figure(figsize=(12,4))
plt.plot(sarimax_pred_df['timestamp'], sarimax_pred_df['actual'], label='Actual', linewidth=1)
plt.plot(sarimax_pred_df['timestamp'], sarimax_pred_df['predicted'], label='Predicted', linewidth=1)
plt.title('Actual vs Predicted (SARIMAX)')
plt.xlabel('Date')
plt.ylabel('Target')
plt.legend()
plt.tight_layout()
plt.show()
```

## 8) Compare Model Performance

```python
comparison = pd.DataFrame([
        {
            'Model': 'LinearRegression',
            'MAE': lin_results['MAE'],
            'MSE': lin_results['MSE'],
            'RMSE': lin_results['RMSE'],
            'R2': lin_results['R2'],
        },
        {
            'Model': 'Statsmodels_SARIMAX',
            'MAE': sarimax_results['MAE'],
            'MSE': sarimax_results['MSE'],
            'RMSE': sarimax_results['RMSE'],
            'R2': sarimax_results['R2'],
        }
    ])
comparison
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=comparison, x='Model', y='RMSE')
plt.title('RMSE Comparison')
plt.tight_layout()
plt.show()
```

## 9) Forecast Future Values
We generate a small future forecast horizon beyond the dataset using SARIMAX's forecasting capability.

```python
future_steps = 30
# Refit SARIMAX on full series for a cleaner 'future' view
y_full = df_feat.sort_values('timestamp')['target'].astype(float)
dates_full = df_feat.sort_values('timestamp')['timestamp']
import statsmodels.api as sm
model_full = sm.tsa.SARIMAX(y_full, order=order, enforce_stationarity=False, enforce_invertibility=False)
res_full = model_full.fit(disp=False)
future_pred = res_full.forecast(steps=future_steps)
future_dates = pd.date_range(dates_full.iloc[-1] + pd.Timedelta(days=1), periods=future_steps, freq='D')
future_df = pd.DataFrame({'timestamp': future_dates, 'predicted': np.asarray(future_pred)})
future_df.head()
```

```python
plt.figure(figsize=(12,4))
plt.plot(df_feat['timestamp'], df_feat['target'], label='Historical', linewidth=1)
plt.plot(future_df['timestamp'], future_df['predicted'], label='Future Forecast', linewidth=2)
plt.title('Future Forecast (SARIMAX)')
plt.xlabel('Date')
plt.ylabel('Target')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Save artifacts
lin_pred_path = os.path.join(ARTIFACTS_DIR, 'linear_regression_predictions.csv')
sarimax_pred_path = os.path.join(ARTIFACTS_DIR, 'sarimax_predictions.csv')
future_pred_path = os.path.join(ARTIFACTS_DIR, 'future_forecast_sarimax.csv')
comparison_path = os.path.join(ARTIFACTS_DIR, 'model_comparison.csv')

lin_pred_df.to_csv(lin_pred_path, index=False)
sarimax_pred_df.to_csv(sarimax_pred_path, index=False)
future_df.to_csv(future_pred_path, index=False)
comparison.to_csv(comparison_path, index=False)

with open(os.path.join(ARTIFACTS_DIR, 'metrics_linear_regression.json'), 'w') as f:
    json.dump({k: lin_results[k] for k in ['MAE','MSE','RMSE','R2']}, f, indent=2)

with open(os.path.join(ARTIFACTS_DIR, 'metrics_sarimax.json'), 'w') as f:
    json.dump({k: sarimax_results[k] for k in ['MAE','MSE','RMSE','R2']}, f, indent=2)

print('Saved artifacts to:', ARTIFACTS_DIR)
```
